# Statistical Classification Log-Loss (Cross Entropy)

This notebook demonstrates the implementation of the Statistical Classification Log-Loss (also known as Cross Entropy) using a healthcare-related dataset. The example involves predicting the likelihood of a patient having a certain condition based on diagnostic test results. The notebook is structured to provide a clear understanding of the algorithm, its implementation, and its significance in classification problems.

```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the Healthcare Dataset
# Example: Diagnostic test results vs. condition presence
# 1 indicates the condition is present, 0 indicates it is not

test_results = np.array([0.1, 0.4, 0.35, 0.8, 0.9, 0.2, 0.5, 0.7, 0.6, 0.3])
condition_present = np.array([0, 0, 0, 1, 1, 0, 1, 1, 1, 0])
```

## Log-Loss Formula

The Log-Loss function measures the performance of a classification model whose output is a probability value between 0 and 1. The formula for Log-Loss is:

$$
\text{Log-Loss} = - \frac{1}{N} \sum_{i=1}^{N} \left( y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right)
$$

Where:
- \( y_i \) is the actual label (0 or 1).
- \( \hat{y}_i \) is the predicted probability of the positive class.
- \( N \) is the total number of samples.

```python
# Implement the Log-Loss Function
def log_loss(y_true, y_pred):
    epsilon = 1e-15  # To avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Calculate Log-Loss for the Healthcare Dataset
loss = log_loss(condition_present, test_results)
print(f"Log-Loss for Healthcare Dataset: {loss}")
```

```python
# Visualize the Results
plt.figure(figsize=(8, 5))
plt.plot(range(len(test_results)), test_results, label="Predicted Probabilities", marker="o")
plt.scatter(range(len(condition_present)), condition_present, color="red", label="Actual Outcomes")
plt.title("Healthcare Dataset: Diagnostic Test Results vs. Condition Presence")
plt.xlabel("Sample Index")
plt.ylabel("Probability / Outcome")
plt.legend()
plt.grid()
plt.show()
```

## Talking Points

1. **Numerical Stability**:
   - The use of `np.clip` in the `log_loss` function ensures numerical stability by preventing logarithmic calculations on values too close to 0 or 1, which could otherwise result in computational errors or infinities.

3. **Visualization**:
   - The visualization highlights the relationship between predicted probabilities and actual outcomes, providing an intuitive understanding of the model's performance.
